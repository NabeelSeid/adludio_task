```python
def filter(text):
    stopwords = ['rt', 'de', ' amp', ' amp ', 'amp ', 'le']
    querywords = text.split()

    resultwords  = [word for word in querywords if word.lower() not in stopwords]
    return ' '.join(resultwords)
```

```python
import re

class TweetFilter():
    
    def __init__(self, df, first_case=None, lockdown_start=None, lockdown_lifted=None, city=None):
        self.df = df
        self.first_case = first_case
        self.lockdown_start = lockdown_start
        self.lockdown_lifted = lockdown_lifted
        self.city = city
    
    def filter_by_date(self, begin_date, end_date):
        return self.df[(self.df.created_at > begin_date) & (self.df.created_at < end_date)]
    
    def filter_by_word_in_hashtag(self, keywords):
        self.df = self.df[
            self.df.text.apply(lambda x: re.search('#[a-zA-Z0-9]*' + '|'.join(keywords), str(x), re.IGNORECASE) != None)]

    def filter_by_keyword(self, keywords):
        self.df = self.df[self.df.text.str.contains('|'.join(keywords), case=False)]
    
    def before_covid(self, first_case):
        self.filter_by_date('2020-01-01', first_case).to_csv('before_covid_{}.csv'.format(self.city))
    
    def during_covid(self, first_case):
        self.filter_by_date(first_case, '2020-10-01').to_csv('during_covid_{}.csv'.format(self.city))
    
    def before_ld(self, lockdown_start):
        self.filter_by_date('2020-01-01', lockdown_start).to_csv('before_lockdown_{}.csv'.format(self.city))
    
    def during_ld(self, lockdown_start, lockdown_lifted):
        self.filter_by_date(lockdown_start, lockdown_lifted).to_csv('during_lockdown_{}.csv'.format(self.city))
    
    def after_ld(self, lockdown_lifted):
        self.filter_by_date(lockdown_lifted, '2020-10-01').to_csv('after_lockdown_{}.csv'.format(self.city))
    
    def filter(self, first_case, lockdown_start, lockdown_lifted):
        self.before_covid(first_case)
        self.during_covid(first_case)
        
        self.before_ld(lockdown_start)
        self.during_ld(lockdown_start, lockdown_lifted)
        self.after_ld(lockdown_lifted)

    def clean(self):
        self.df = CleanTweet(self.df).clean()
```
